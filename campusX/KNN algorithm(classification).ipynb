```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification


```

## lets create a dataset first

```python
# Reproducibility
np.random.seed(42)

# Generate data
X,y = make_classification(n_samples = 100,n_features = 2,n_redundant = 0,n_classes = 2,n_clusters_per_class = 1,random_state=41,hypercube = False,class_sep=2)
```

```python
print(X)
```

```python
print(y)
```

```python
## lets consider a random point which we need to classify
point = [np.random.rand(),np.random.rand()]
print(point)
```

## lets plot the points and the target point

```python
plt.figure(figsize = (6,4))
sns.scatterplot(x = X[:,0],y = X[:,1],hue = y)
sns.scatterplot(x = [point[0]],y = [point[1]],color = "red")
```

```python

result_list = []
## knn algo
for i in range(len(X)):
  #now we need to find the distance between the random point and the target point
  p1 = X[i]
  p2 = point
  distance = np.sqrt(np.sum((p1 - p2) ** 2))
  answer = [X[i],y[i],distance]

  #appending the result in result_list
  result_list.append(answer)


```

```python
print(result_list)
```

```python
# now we need to sort the list obtained above
answer_df = pd.DataFrame(result_list,columns = ["relative point","relative point label","distance"])

answer_df.head()
```

```python
## now we will sort the above data
answer_df = answer_df.sort_values(by = "distance")
answer_df.head()
```

```python
# let epochs be  = 10
answer_df = answer_df.head(10)
answer_df.groupby("relative point label")["relative point label"].count()
```
