```python
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
X,y = datasets.make_blobs(
    n_samples = 100,
    n_features = 2,
    centers = 2,
    random_state = 50,
    cluster_std = 0.8
)
```

## so first we will code hard-margin svm

```python
print(y)
```

```python
## now lets plot the vectors
plt.figure(figsize = (6,4))
plt.scatter(x = X[:,0],y = X[:,1],c =y)
```

```python
## now we will train the model
model = SVC(kernel= "linear",C=10)
model.fit(X,y)

model1 = SVC(kernel= "linear",C=0.002)
model1.fit(X,y)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize = (12,4))
ax[0].set_xlim(-10,4)
ax[0].set_ylim(-10,2)
ax[1].set_xlim(-10,4)
ax[1].set_ylim(-10,2)
ax[0].scatter(X[:,0],X[:,1],c=y)
ax[1].scatter(X[:,0],X[:,1],c=y)

## now we will create mesh grids
xx0,yy0 = np.meshgrid(
    np.linspace(ax[0].get_xlim()[0],ax[0].get_xlim()[1],100),
    np.linspace(ax[0].get_ylim()[0],ax[0].get_ylim()[1],100)
)

xx1,yy1 = np.meshgrid(
    np.linspace(ax[1].get_xlim()[0],ax[1].get_xlim()[1],100),
    np.linspace(ax[1].get_ylim()[0],ax[1].get_ylim()[1],100)
)

## to show decision boundary
fake_points1 = np.c_[xx0.ravel(), yy0.ravel()]
fake_point_prediction = model.predict(fake_points1).reshape(xx0.shape)

ax[0].contourf(xx0, yy0, fake_point_prediction, alpha=0.3)

#now we will show the lines
fake_points2 = np.c_[xx1.ravel(), yy1.ravel()]
fake_point_decision_values = model.decision_function(fake_points2).reshape(xx1.shape)
ax[1].contour(xx1,yy1,fake_point_decision_values,alpha = 0.3,levels = [-1,0,1],colors = "black",linestyles=['--', '-', '--'])

ax[0].set_title("decision boundary")
ax[1].set_title("the hyperplanes")




```
